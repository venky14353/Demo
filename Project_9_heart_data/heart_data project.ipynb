```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# importing heart data csv file 
heart_data = pd.read_csv("/content/heart_disease_data.csv")
heart_data
```

```python
# here 1 is heart patient and 0 is healthy person
# splitting the data into features and target columns
X = heart_data.drop(columns = 'target', axis=1)
Y = heart_data['target']
print(X)
print(Y)
```

```python
# splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
print(X.shape,X_train.shape, X_test.shape)
```

```python
# now we train the model with logistic regression
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
# model evaluation - accuracy score
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print(" Accuracy on training data : ", training_data_accuracy)
```

```python
# accuracy score on testing data
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy on testing data : ", training_data_accuracy)
```
