```python
# importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# import dataset and data preprocessing
parkinsons_dataset = pd.read_csv('/content/parkinsons.csv')
parkinsons_dataset
```

```python
# splitting x and y
x = parkinsons_dataset.drop(columns=['name','status'], axis=1)
y = parkinsons_dataset['status']
```

```python
# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
# data standardization
scalar = StandardScaler()
scalar.fit(x_train)
```

```python
x_train = scalar.transform(x_train)
x_test = scalar.transform(x_test)
```

```python
# model training : support vector machine classifier
model = svm.SVC(kernel = 'linear')
model.fit(x_train, y_train)
```

```python
# model evaluation : accuracy score
x_train_pred = model.predict(x_train)
training_data_acc = accuracy_score(y_train, x_train_pred)
training_data_acc
```

```python
x_test_pred = model.predict(x_test)
test_data_acc = accuracy_score(y_test, x_test_pred)
test_data_acc
```

```python
# making a predictive model
input_data = (199.22800,209.51200,192.09100,0.00241,0.00001,0.00134,0.00138,0.00402,0.01015,0.08900,0.00504,0.00641,0.00762,0.01513,0.00167,30.94000,0.432439,0.742055,-7.682587,0.173319,2.103106,0.068501)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
std_data = scalar.transform(input_data_reshaped)
prediction = model.predict(std_data)

print(prediction)
if (prediction[0]==0):
  print("the person is healthy",prediction)
else:
  print("the person have parkersons disease")
```
