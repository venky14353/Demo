```python
# importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# dataset loading and data preprocessing
credit_card_data = pd.read_csv(r"C:\Users\my hp\Downloads\Project_10_Credit_Card_Fraud_Detection\creditcard.csv")
```

```python
# dataset higlhly unbalanced : 0 for normal transaction, 1 for fraudulent transaction
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# checking the statistical measurements of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
# Under sampling data, building a sample dataset containing similar legit and fraud transactions
# no. of fraudulent transactions = 492
legit_sample = legit.sample(n=492)
```

```python
# concatenating the datasets
new_dataset = pd.concat([legit_sample,fraud],axis=0)
new_dataset
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting the data into features and targets
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
X, Y
```

```python
# splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
# model training : Logistic Regression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# model evaluation - prediction on training data
X_train_pred = model.predict(X_train)
# accuracy on training data
training_data_accuracy = accuracy_score(Y_train, X_train_pred)
training_data_accuracy
```

```python
# prediction on testing data
X_test_pred = model.predict(X_test)
# accuracy on testing data
test_data_accuracy = accuracy_score(Y_test, X_test_pred)
test_data_accuracy
```

```python
#classification report
from sklearn import metrics
Y_pred = model.predict(X_test)
cr = metrics.classification_report(Y_test, Y_pred)
cr
```

```python
score_1 = metrics.r2_score(Y_pred,Y_test)
score_1
```

```python
mae = metrics.mean_absolute_error(Y_pred,Y_test)
mae
```

```python
cm = metrics.confusion_matrix(Y_pred,Y_test)
cm
```
